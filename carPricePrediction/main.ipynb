```python
import sklearn
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
# Random forest does not need scaling as it uses decision tree
from sklearn.ensemble import RandomForestRegressor
rf_random = RandomForestRegressor()
# Random Forest has lots of parameters, so hyperparameter tuning is needed
```

```python
### Hyperparameters - important to do hyperparameter tuning to get better accuracy
import numpy as np
n_estimators = [int(x) for x in np.linspace(start=100, stop=1200, num=12)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(start=5, stop=30, num=6)]
min_samples_split = [2,5,10,15,100]
min_samples_leaf = [1,2,5,10]
print(n_estimators)
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Can also use GridSearchCV but RandomizedSearchCV is faster
```

```python
#Creating random grid
random_grid = {
    'n_estimators': n_estimators,
    'max_features': max_features,
    'max_depth': max_depth,
    'min_samples_split': min_samples_split,
    'min_samples_leaf': min_samples_leaf
}
print(random_grid)
```

```python
# First create the base model to tune. This is a default model and it will be trained based on different parameters
# understand the meaning of all the parameters present below
rf = RandomForestRegressor()
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, scoring='neg_mean_squared_error', n_iter=10, cv=5, verbose=2, random_state=42, n_jobs=1)
```

```python
rf_random.best_params_
```

```python
predictions = rf_random.predict(X_test)
```

```python
predictions
```

```python
sns.displot(y_test - predictions)
# A normal distribution is shown in the created graph, which shows good results
```

```python
plt.scatter(y_test, predictions)
# Linear predictions means good accuracy
```
